```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('BX-Books.csv', sep=';', error_bad_lines=False, encoding='latin1')
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.head()
```

```python
# Lets rename some wierd column name
books.rename(columns={"Book-Title": 'title',
                      'Book-Author': 'author',
                      "Year-Of-Publication": 'year',
                      "Publisher": "publisher",
                      "Image-URL-L": "image_url"},inplace=True)
```

```python
books.head()
```

```python
users = pd.read_csv('BX-Users.csv', sep=';', error_bad_lines=False, encoding='latin1')
```

```python
users.head()
```

```python
users.shape
```

```python
# Lets rename some wierd columns name
users.rename(columns={"User-ID": 'user_id',
                      "Location": 'location',
                      "Age": 'age'}, inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding='latin1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
# Lets rename some wierd columns name
ratings.rename(columns={"User-ID": 'user_id',
                      "Book-Rating": 'rating'}, inplace=True)
```

```python
ratings.head()
```

```python
print(books.shape, users.shape, ratings.shape, sep='\n')
```

```python
ratings['user_id'].value_counts()
```

```python
# Lets store users who had at least rataed more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
# Now join ratings with books

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
# Lets take those books which got at least 50 rating of user

final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
# Lets drop the dupliactes
final_rating.drop_duplicates(['user_id', 'title'],inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched '{book_name}'\n")
                print("The suggestion books are: \n")
            else:
                print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```

```python
book_name = '2nd Chance'
recommend_book(book_name)
```
